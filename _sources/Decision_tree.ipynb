## 2. Entropi Dataset Awal (Parent Node)

jumlah data untuk masing-masing kelas:

* Kelas **G**: 6 data
  (baris: 1, 2, 3, 5, 12, 15)
* Kelas **H**: 9 data
  (baris: 4, 6, 7, 8, 9, 10, 11, 13, 14)

Total data = 15
Maka probabilitasnya:

* $p_G = \frac{6}{15} = 0.4$
* $p_H = \frac{9}{15} = 0.6$

Sekarang kita hitung **Entropi** untuk parent node (dataset keseluruhan):



###  **Perhitungan Entropi:**

$$
\begin{aligned}
\text{Entropy}(S) &= -\left( p_G \cdot \log_2(p_G) + p_H \cdot \log_2(p_H) \right) \\
&= -\left( 0.4 \cdot \log_2(0.4) + 0.6 \cdot \log_2(0.6) \right) \\
&= -\left( 0.4 \cdot (-1.322) + 0.6 \cdot (-0.737) \right) \\
&= -(-0.5288 - 0.4422) \\
&= 0.971
\end{aligned}
$$

$$

###  Entropi Subset 2 (`Panjang > 185`):

$$
\begin{aligned}
p_G &= \frac{6}{7}, \quad p_H = \frac{1}{7} \\
\text{Entropy} &= -\left( \frac{6}{7} \log_2 \frac{6}{7} + \frac{1}{7} \log_2 \frac{1}{7} \right) \\
&= -\left( 0.857 \cdot (-0.222) + 0.143 \cdot (-2.807) \right) \\
&= 0.190 + 0.402 = \mathbf{0.592}
\end{aligned}
$$



### **Info Gain dari Split “Panjang <= 185”**

$$
\begin{aligned}
\text{Entropy Total (Parent)} &= 0.971 \\
\text{Rata-rata Entropy Child} &= \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0.592 \\
&= 0 + 0.276 \\
\text{Information Gain} &= 0.971 - 0.276 = {0.695}
\end{aligned}
$$



## 4. Split Berdasarkan Fitur: "Berat Tubuh (kg)"

###  Filter data dengan `Berat <= 155` (dari dataset baru):

| No | Berat | Kelas |
| -- | ----- | ----- |

Total = 8

###  Entropi untuk subset `Berat <= 155`:

$$
\begin{aligned}
p_G &= \frac{2}{8} = 0.25, \quad p_H = \frac{6}{8} = 0.75 \\
\text{Entropy} &= -\left( 0.25 \log_2 0.25 + 0.75 \log_2 0.75 \right) \\
&= -\left( 0.25 \cdot -2 + 0.75 \cdot -0.415 \right) \\
&= 0.5 + 0.311 = \boxed{0.811}
\end{aligned}
$$



### Data dengan `Berat > 155`:

| No | Berat | Kelas |
| -- | ----- | ----- |
| 1  | 160   | G     |

Total = 7
Kelas G = 4 (No: 1, 2, 12, 15)
Kelas H = 3 (No: 4, 6, 13)



###  Entropi untuk subset `Berat > 155`:

$$
\begin{aligned}
p_G &= \frac{4}{7} \approx 0.571, \quad p_H = \frac{3}{7} \approx 0.429 \\
\text{Entropy} &= -\left( 0.571 \cdot \log_2 0.571 + 0.429 \cdot \log_2 0.429 \right) \\
&= -(0.571 \cdot -0.807 + 0.429 \cdot -1.222) \\
&= 0.461 + 0.524 = {0.985}
\end{aligned}
$$



### **Information Gain dari Split “Berat <= 155”**

$$
\begin{aligned}
\text{Entropy Parent} &= 0.971 \\
\text{Entropy Rata-Rata} &= \frac{8}{15} \cdot 0.811 + \frac{7}{15} \cdot 0.985 \\
&= 0.4325 + 0.4597 = 0.892 \\
\text{Information Gain} &= 0.971 - 0.892 = {0.079}
\end{aligned}
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Baru ===
data = {
    "Panjang (cm)": [180, 195, 200, 188, 190, 175, 165, 160, 170, 168, 160, 198, 185, 162, 196],
    "Berat (kg)": [160, 170, 150, 165, 155, 158, 150, 145, 140, 148, 139, 162, 168, 142, 158],
    "Kelas": ['G', 'G', 'G', 'H', 'G', 'H', 'H', 'H', 'H', 'H', 'H', 'G', 'H', 'H', 'G']
}

df = pd.DataFrame(data)

# === Encode label G/H jadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # G = 0, H = 1 (tergantung urutan alfabet)

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Tampilkan aturan pohon (rules) ===
print("\n=== Aturan Pohon Keputusan ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# === Visualisasi Pohon Keputusan ===
plt.figure(figsize=(12, 6))
plot_tree(clf,
          feature_names=["Panjang (cm)", "Berat (kg)"],
          class_names=le.classes_,  # ['G', 'H']
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Visualisasi Pohon Keputusan")
plt.show()
```

## Visualisasi Splitting Setelah ada 2 Data Baru

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Sesuai (G/H dan Nilai Baru) ===
data = {
    "Panjang (cm)": [180, 195, 200, 188, 190, 175, 165, 160, 170, 168, 160, 198, 185, 162, 196],
    "Berat (kg)": [160, 170, 150, 165, 155, 158, 150, 145, 140, 148, 139, 162, 168, 142, 158],
    "Kelas": ['G', 'G', 'G', 'H', 'G', 'H', 'H', 'H', 'H', 'H', 'H', 'G', 'H', 'H', 'G']
}

df = pd.DataFrame(data)

# === Encode label G/H menjadi 0/1 ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # G=0, H=1 (karena urutan alfabet)

# === Fitur dan Target ===
X = df[["Panjang (cm)", "Berat (kg)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X, y)

# === Prediksi 2 Data Baru ===
new_data = pd.DataFrame({
    "Panjang (cm)": [192, 163],   # nilai baru bisa kamu sesuaikan
    "Berat (kg)": [167, 149]
})
preds = clf.predict(new_data)
predicted_labels = le.inverse_transform(preds)

# === Visualisasi Decision Boundary ===
x_min, x_max = X["Panjang (cm)"].min() - 5, X["Panjang (cm)"].max() + 5
y_min, y_max = X["Berat (kg)"].min() - 5, X["Berat (kg)"].max() + 5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 1),
                     np.arange(y_min, y_max, 1))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X["Panjang (cm)"], X["Berat (kg)"], c=y, cmap=plt.cm.coolwarm, edgecolor="k", s=60, label="Data Latih")
plt.scatter(new_data["Panjang (cm)"], new_data["Berat (kg)"], c="yellow", edgecolor="black", s=100, label="Data Baru")
for i, label in enumerate(predicted_labels):
    plt.text(new_data.iloc[i, 0] + 0.5, new_data.iloc[i, 1], label, fontsize=10, weight='bold')

plt.xlabel("Panjang (cm)")
plt.ylabel("Berat (kg)")
plt.title("Decision Boundary dengan Data Baru")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Output Label Prediksi ===
print("Prediksi Kelas untuk Data Baru:", predicted_labels)
```
